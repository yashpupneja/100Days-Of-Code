Let us assume there are two sets:
    
1. X1: [x1',x2'.......xn'] with n observations

2. X2: m observations

is the distribution of X1 same as X2 ?

https://en.wikipedia.org/wiki/Kolmogorov%E2%80%93Smirnov_test
    
Two sample K-S test topic

Red line is CDF of X1;
Blue line is CDF of X2;

Null Hypothesis : X1 and x2 have same distribution

sup means the maximal difference value accord to formula of Dn,m

(p is denoted by alpha here )

if n=1000, m=5000 and alpha=5%, then acc to formula Dn,m>0.047, Hence reject the null hypothesis at 5% significant value

if n=50 , m=30 and alpha =5%, then acc to formula Dn,m>0.31, Hence reject the null hypothesis

```python
import numpy as np
import seaborn as sns 
from scipy import stats
import matplotlib.pyplot as plt
```

```python
#generate a gaussian random variable X
x=stats.norm.rvs(size=1000); #generating X~N(0,1) with 1000 such points
sns.set_style('whitegrid')
sns.kdeplot(np.array(x), bw=0.5) #kernel density distribution
plt.show()
```

```python
#since pvalue is very high hence  the distribution is normal distributed
stats.kstest(x,'norm')   #to see whether the the distribution is(comparing) normal distribution or not
```

```python
#y-continuous uniform distribution(0,1)
y=np.random.uniform(0,1,10000);
sns.kdeplot(np.array(y),bw=0.1)
plt.show()
```

```python
stats.kstest(y,'norm') #not gaussian distribution
```
