## Turtles

```python
import turtle
```

```python
window=turtle.Screen()
window.bgcolor("red")
```

```python
stake=turtle.Turtle() #Turtle is a class
```

```python
stake.color("black") #yellow
stake.shape("turtle") #arrow
stake.forward(90)
stake.right(90)
stake.forward(90)
stake.right(90)
stake.forward(90)
stake.right(90)
stake.forward(90)
window.exitonclick()
```

```python
import turtle
window=turtle.Screen()
window.bgcolor("red")
sk=turtle.Turtle()
sk.color("black") #yellow
sk.shape("turtle")
i=0
while i<4:
    sk.forward(90)
    sk.right(90)
    i+=1
window.exitonclick()
```

```python
import turtle
window=turtle.Screen()
window.bgcolor("red")
s=turtle.Turtle()
s.color("black")
s.shape("arrow")
d=250
angle=90
for i in range(1,251):
    s.forward(d)
    s.left(angle)
    d-=1
window.exitonclick()
```

```python
import turtle
window=turtle.Screen()
window.bgcolor("yellow")
y=turtle.Turtle()
y.color("red")
y.shape("arrow")
y.penup()
y.goto(-500,130)
y.pendown()
degree=45
#Y
y.right(degree)
y.forward(90)
y.left(90)
y.forward(90)
y.backward(90)
y.right(135)
y.forward(90)
#A
y.penup()
y.goto(-359.36,-23.64)
y.pendown()
y.left(165)
y.forward(150)
y.pos()
y.right(145)
y.forward(150)
y.right(180)
y.forward(75)
y.left(69)
y.forward(45)
y.penup()

#S
y.goto(-230,120)
y.pendown()
y.circle(35,180)
y.forward(2)
y.penup()
y.goto(-250,-23.64)
y.pendown()
y.circle(39,180)

#H
y.penup()
y.goto(-180,-23.64)
y.pendown()
y.right(91)
y.forward(150)
y.backward(75)
y.right(90)
y.forward(80)
y.left(90)
y.forward(75)
y.backward(150)

#A
y.penup()
y.goto(-80,-23.64)
y.pendown()
y.right(15)
y.forward(150)
y.right(145)
y.forward(150)
y.right(180)
y.forward(75)
y.left(69)
y.forward(45)

#S
y.penup()
y.goto(60,120)
y.pendown()
y.circle(32,180)
y.forward(2)
y.penup()
y.goto(55,-23.64)
y.pendown()
y.circle(40,180)

#W
y.penup()
y.goto(100,120)
y.pendown()
y.right(255)
y.forward(139)
y.left(150)
y.forward(125)
y.right(140)
y.forward(125)
y.left(150)
y.forward(125)
y.penup()

#I
y.goto(280,120)
y.pendown()
y.right(82)
y.forward(90)
y.backward(45)
y.right(90)
y.forward(120)
y.right(90)
y.forward(45)
y.backward(90)
window.exitonclick()
```
